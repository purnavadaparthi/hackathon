```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace this with your actual data)
ratings_data = {'userid': [1, 2, 3, 4, 5],
                'movieid': [301, 301, 301, 301, 301],
                'rating': [4, 5, 4, 3, 5],
                'timestamp': [1609459200, 1609545600, 1609632000, 1609718400, 1609804800]}

ratings_df = pd.DataFrame(ratings_data)

# Filter ratings for the movie "Fight Club (1999)"
movie_title = "Fight Club (1999)"
movie_id = 301

movie_ratings = ratings_df[ratings_df['movieid'] == movie_id]['rating']

# Plot histogram
plt.hist(movie_ratings, bins=5, edgecolor='black')
plt.title(f'Distribution of User Ratings for "{movie_title}"')
plt.xlabel('User Ratings')
plt.ylabel('Frequency')
plt.show()
```
